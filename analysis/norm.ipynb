# Visualization of the used normal distribution for the patience generation

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pylab as plt
import matplotlib as mpl
```

```python
# CONFIGURATION
MEAN = 15
DEVIATION = 20
```

```python
STANDARD_DEVIATION = DEVIATION / 3

MIN = MEAN - DEVIATION
MAX = MEAN + DEVIATION

X_MIN = MIN - 5
X_MAX = MEAN * 2 - X_MIN

x = np.linspace(X_MIN, X_MAX, 200)
y = norm.pdf(x, MEAN, STANDARD_DEVIATION)
```

```python
from matplotlib.pylab import rc
rc('font',**{'family':'sans-serif','sans-serif':['Bahnschrift']})
rc('text', usetex=True)
mpl.rcParams['pdf.fonttype'] = 42

fig, axes = plt.subplots(1, figsize=(14, 7))

axes.set_title(f'Normalverteilung zum Generieren der Geduld\nmit $\mu = {MEAN}$ und $\sigma = {round(STANDARD_DEVIATION, 3)}$ (max. Abweichung = ${DEVIATION}$)', fontsize=28, color=(0.1, 0.1, 0.1))
axes.title.set_position([.5, 1.05])



axes.plot(x, y, color=(1.0, 0.2, 0.4, 1.0), linewidth=3)

fill_x = np.linspace(MIN, MAX, 200)
fill_y = norm.pdf(fill_x, MEAN, STANDARD_DEVIATION)
axes.fill_between(fill_x, 0, fill_y, color=(1.0, 0.2, 0.4), alpha=0.2)

axes.set_ylabel('Wahrscheinlichkeit $p(x)$', fontsize=20, color='slategrey')
axes.set_xlabel('Geduld', fontsize=20, color='slategrey')

axes.grid(color='grey', linestyle='dotted', linewidth=2, alpha=0.1)

axes.margins(0.02)

axes.set_xlim(X_MIN, X_MAX)
axes.set_ylim(0.0, max(y) * 1.1)

plt.setp(axes.spines.values(), color='slategrey')
plt.setp([axes.get_xticklines(), axes.get_yticklines()], color='slategrey')
axes.tick_params(axis='both', which='major', labelsize=14)
axes.tick_params(axis='both', which='minor', labelsize=12)

fig.subplots_adjust(top=0.8)

fig.savefig('out/patience_norm_dist.pdf', transparent=True)  
```
